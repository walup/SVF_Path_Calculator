```python
from Panoramic2FishEye import Panoramic2FishEye
import matplotlib.pyplot as plt
```

```python
#Initialize the fish image converter
fishImageWidth = 1000
fishImageHeight = 1000

fishImgConverter = Panoramic2FishEye(fishImageWidth, fishImageHeight)
#Open the panoramic image, in this case with path 
fishImgConverter.setPanoramicWithPath("test_image.png")
#Convert to fish eye image

plt.figure()
plt.imshow(fishImgConverter.getPanoramicImage(),cmap = 'gray')
```

```python
#Get the fish image
radiusFraction = 1
fishImage = fishImgConverter.convert2FishEye(radiusFraction)

plt.figure()
plt.imshow(fishImage)
```
