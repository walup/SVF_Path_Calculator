```python
from PIL import Image
import matplotlib.pyplot as plt

img = np.asarray(Image.open('test_2.png'))[:,:,0:3]
segmentator = SemanticSegmentation()
maskedImage = segmentator.segmentImage(img)

fig, ax = plt.subplots(2,1, figsize = (20, 10))
ax[0].imshow(img)
ax[0].set_title("Non-masked image")
ax[1].imshow(maskedImage)
ax[1].set_title("Masked image")
```
